```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
# Load the dataset
census = pd.read_csv('Data/census_data.csv')
census.head()
```

```python
# Preprocessing
# Assuming 'income_bracket' is the target variable
encoder = LabelEncoder()
census['income'] = encoder.fit_transform(census['income_bracket'])

# Identify the categorical columns
cat_cols = ['workclass', 'education', 'marital_status', 'occupation', 'relationship', 'race', 'gender', 'native_country']

# Perform one-hot encoding for categorical columns
census = pd.get_dummies(census, columns=cat_cols, drop_first=True)

# Define features (X) and target (y)
X = census.drop(['income_bracket', 'income'], axis=1)
y = census['income']
```

```python
census.head()
# View columns 20 to 29 in the census DataFrame
#census.iloc[:, 20:30]
# Display all column names in the census DataFrame
# Display all column names in the census DataFrame without truncation
print(census.columns.tolist())

```

```python


# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)

# Create a Random Forest Classifier
rf_classifier = RandomForestClassifier(random_state=101)

# Fit the model to the training data
rf_classifier.fit(x_train, y_train)

# Make predictions on the test data
y_pred_rf = rf_classifier.predict(x_test)

# Calculate accuracy
accuracy_rf = accuracy_score(y_test, y_pred_rf)
print("Random Forest Classifier Accuracy:", accuracy_rf)

# Generate a classification report
report_rf = classification_report(y_test, y_pred_rf)
print("Classification Report for Random Forest Classifier:\n", report_rf)
```

```python
import joblib

# Save the model to a file
model_filename = 'random_forest_model.joblib'
joblib.dump(rf_classifier, model_filename)

print("Model saved to", model_filename)
```
